```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fnt_size = 14
plt.rc('font', size=fnt_size)          # controls default text sizes
plt.rc('axes', titlesize=fnt_size)     # fontsize of the axes title
plt.rc('axes', labelsize=fnt_size)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=fnt_size)    # fontsize of the tick labels
plt.rc('ytick', labelsize=fnt_size)    # fontsize of the tick labels
plt.rc('legend', fontsize=fnt_size)    # legend fontsize
plt.rc('figure', titlesize=fnt_size)  # fontsize of the figure title
```

```python
def tsigmoid(x, w):
    return 1 - math.log(w+1,2) / (1 + math.exp(x))  # 0<w<1
```

```python
x = np.linspace(-5,5, 100)
df = pd.DataFrame()
df['x'] = x
df['w0'] = list(map(lambda z: tsigmoid(z,0),x))
df['w03'] = list(map(lambda z: tsigmoid(z,0.3),x))
df['w05'] = list(map(lambda z: tsigmoid(z,0.5),x))
df['w07'] = list(map(lambda z: tsigmoid(z,0.7),x))
df['w1'] = list(map(lambda z: tsigmoid(z,1),x))
```

```python
#define colors to use
col1 = 'Blues'

# dfinterpolatelast = dfinterpolation_summary[dfinterpolation_summary['interpolate_position'] == 'last'].sort_values("w")
# dfinterpolateall = dfinterpolation_summary[dfinterpolation_summary['interpolate_position'] == 'all'].sort_values("w")

#define subplots
fig,ax = plt.subplots()

ax.plot(df["x"], df['w0'],color = "midnightblue") #, edgecolors='grey')
ax.plot(df["x"], df['w03'],color = "royalblue")
ax.plot(df["x"], df['w05'],color = "steelblue")
ax.plot(df["x"], df['w07'],color = "deepskyblue")
ax.plot(df["x"], df['w1'],color = "lightskyblue")


#ax.plot(df_FT["Best NFaiRR_cutoff_10"], df_FT['Best MRR@10'],c=df_FT['bias_regul_coeff'], cmap=col3, marker = "o",vmin=-20., vmax=40,s= 60) #, edgecolors='grey')

ax.set_ylim(ymin=0)
ax.yaxis.set_ticks(np.arange(0, 1, 0.2))
#add x-axis label
ax.set_xlabel('x')
#add y-axis label
ax.set_ylabel('activation')
ax.grid(axis='x')

ax.legend(['$\omega$ = 0', '$\omega$ = 0.3', '$\omega$ = 0.5', '$\omega$ = 0.7','$\omega$ = 1'], loc = 'best')

plt.savefig("/mnt/c/Users/Cornelia/Documents/AI/MasterThesis/Results/tsigmoid.pdf", format="pdf", bbox_inches="tight", dpi = 500)
```
